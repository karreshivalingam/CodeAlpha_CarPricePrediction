```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/car data.csv")
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["Selling_Price"], bins=20)
plt.title("Selling Price Distribution")
plt.xlabel("Selling Price (Lakhs)")
plt.ylabel("Count")
plt.show()
```

```python
numeric_cols = ["Year", "Selling_Price", "Present_Price", "Driven_kms", "Owner"]
corr = df[numeric_cols].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Present_Price"], df["Selling_Price"])
plt.title("Selling Price vs Present Price")
plt.xlabel("Present Price")
plt.ylabel("Selling Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Driven_kms"], df["Selling_Price"])
plt.title("Selling Price vs Driven kms")
plt.xlabel("Driven kms")
plt.ylabel("Selling Price")
plt.show()
```
